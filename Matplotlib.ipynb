```python
x = np.array([0,1,2,3,4,5])
y = np.array([0,8,12,20,26,38])
```

```python
font1 = {
  'family': 'serif',
  'color': 'blue',
  'size': 20
}
```

```python
font2 = {
  'family': 'serif',
  'color': 'yellow',
  'size': 15
}
```

```python
plt.plot(x,y)
plt.xlabel('Overs', fontdict= font2)
plt.ylabel('Runs', fontdict=font1)
plt.title('IPL', loc='left')
plt.grid(True)
plt.show()
```

```python
x= np.array([0,1,2,3,4,5])
y= np.array([0,8,12,20,26,38])
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.xlabel('Overs', fontdict=font2)
plt.ylabel('Runs', fontdict=font1)
plt.title('IPL Plot 1', loc='left')
plt.grid(True)

x = np.array([0,1,2,3,4,5])
y = np.array([0,8,12,20,26,38])
plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.xlabel('Runs', fontdict=font2)
plt.ylabel('Overs', fontdict=font1)
plt.title('IPL Plot 2', loc='left')
plt.suptitle("Cricket")
plt.grid(True)

plt.tight_layout()
plt.show()
```

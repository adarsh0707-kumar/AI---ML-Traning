```python
df['Gender'] = le.fit_transform(df['Gender'])
df['Married'] = le.fit_transform(df['Married'])
df['Education'] = le.fit_transform(df['Education'])
df['Self_Employed'] = le.fit_transform(df['Self_Employed'])
df['Property_Area'] = le.fit_transform(df['Property_Area'])
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])
df['Dependents'] = le.fit_transform(df['Dependents'])
```

```python
df['Gender'].fillna(df['Gender'].mean(), inplace=True)
df['Dependents'].fillna(df['Dependents'].mean(), inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mean(), inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean(), inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mean(), inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
X = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
model_rf = RandomForestClassifier()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.linear_model import LogisticRegression
model_Lr = LogisticRegression()
```

```python
model_Lr.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model_Lr.predict(X_test))
```

```python
confusion_matrix(y_test, model_Lr.predict(X_test))
```

```python
print(classification_report(y_test, model_Lr.predict(X_test)))
```

```python
model_knn = KNeighborsClassifier()
```

```python
model_knn.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model_knn.predict(X_test))
```

```python
confusion_matrix(y_test, model_knn.predict(X_test))
```

```python
print(classification_report(y_test, model_knn.predict(X_test)))
```

```python
model_rf.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model_rf.predict(X_test))
```

```python
confusion_matrix(y_test, model_rf.predict(X_test))
```

```python
print(classification_report(y_test, model_rf.predict(X_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier()
```

```python
model_dt.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model_dt.predict(X_test))
```

```python
confusion_matrix(y_test, model_dt.predict(X_test))
```

```python
print(classification_report(y_test, model_dt.predict(X_test)))
```

```python
model_vc = VotingClassifier(estimators=[('rf', model_rf), ('lr', model_Lr), ('knn', model_knn), ('dt', model_dt)], voting='hard')


```

```python
model_vc.fit(X_train, y_train)
```

```python
accuracy_score(y_test, model_vc.predict(X_test))
```

```python
confusion_matrix(y_test, model_vc.predict(X_test))
```

```python
print(classification_report(y_test, model_vc.predict(X_test)))
```
